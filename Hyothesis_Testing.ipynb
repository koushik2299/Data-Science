# ChiSquare

It is a hypothesis testing where we test the validity of hypothsis if it talks about frequencies of categorical varibles.



```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image
img = mpimg.imread('/content/chisquare.png')

# Display the image
plt.imshow(img)
plt.show()
```

```python
import scipy.stats as stat
import numpy as np
```

```python
expected_data=[8,6,7,9,6,9,7]
observed_data=[7,8,6,9,9,6,7]
```

```python
sum(expected_data),sum(observed_data)
```

```python
chi_square_test_stat,p_value = stat.chisquare(observed_data,expected_data)
print("X Square ",chi_square_test_stat)
print("P-value ",p_value)
```

```python
siginificance = 0.05
dof=len(observed_data)-1
critical_value=stat.chi2.ppf(1-siginificance,dof)
print(critical_value)
```

```python
if chi_square_test_stat>critical_value:
  print("Null Hypothesis is Rejected")
else:
  print("Null Hypothesis is Accepted")
```
